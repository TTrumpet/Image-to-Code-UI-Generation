```python
import os
import base64
from openai import OpenAI

# --- 1Ô∏è‚É£ Initialize client ---
client = OpenAI(api_key="")  # <-- Replace or set via env var

# --- 2Ô∏è‚É£ Folder containing wireframe images ---
folder_path = r"C:\Users\fa076154\Desktop\CAP6411\Inference\inference_images"  # <-- Change this to your folder path
output_dir = "generated_html_gpt"
os.makedirs(output_dir, exist_ok=True)

# --- 3Ô∏è‚É£ Supported image formats ---
valid_exts = (".png", ".jpg", ".jpeg")

# --- 4Ô∏è‚É£ Loop through all images in the folder ---
for filename in os.listdir(folder_path):
    if not filename.lower().endswith(valid_exts):
        continue

    image_path = os.path.join(folder_path, filename)
    print(f"‚öôÔ∏è Processing: {filename}")

    # Read and encode image
    with open(image_path, "rb") as f:
        image_bytes = f.read()
    image_b64 = base64.b64encode(image_bytes).decode("utf-8")

    # Define prompt
    prompt = (
    "You are an expert front-end developer and UI designer. "
    "Given ONLY the provided wireframe image, generate a complete, minimal, and responsive HTML5 page that faithfully recreates the layout and color scheme.\n\n"
    "Requirements:\n"
    "- Reproduce the visual hierarchy and spatial layout exactly as shown in the wireframe (headers, hero sections, cards, sidebars, buttons, forms, footers, etc.).\n"
    "- Use semantic HTML tags (e.g., <header>, <nav>, <main>, <section>, <article>, <footer>).\n"
    "- Include a single <style> block within the HTML (no external files or JS).\n"
    "- CSS must:\n"
    "    ‚Ä¢ Maintain correct element alignment, spacing, and proportions using flexbox or CSS grid.\n"
    "    ‚Ä¢ Approximate the wireframe‚Äôs color palette (detect and apply primary, secondary, and accent colors where visible).\n"
    "- Use placeholder text for any text elements (e.g., 'Heading', 'Lorem ipsum', 'Button').\n"
    "- Do NOT include explanations, comments, or markdown ‚Äî output ONLY a valid, standalone HTML document.\n\n"
    "Focus on clean structure, proper indentation, and accurate visual representation."
)

    # Send request
    try:
        response = client.chat.completions.create(
            model="gpt-4o-mini",
            messages=[
                {"role": "system", "content": "You convert UI wireframes into production-ready HTML."},
                {
                    "role": "user",
                    "content": [
                        {"type": "text", "text": prompt},
                        {"type": "image_url", "image_url": {"url": f"data:image/png;base64,{image_b64}"}}
                    ]
                }
            ]
        )

        html_output = response.choices[0].message.content.strip()

        # Save output HTML
        output_path = os.path.join(output_dir, f"{os.path.splitext(filename)[0]}.html")
        with open(output_path, "w", encoding="utf-8") as f:
            f.write(html_output)

        print(f"‚úÖ Generated HTML saved to: {output_path}")

    except Exception as e:
        print(f"‚ùå Error processing {filename}: {e}")

print("\nüéâ All wireframes processed successfully!")
```
